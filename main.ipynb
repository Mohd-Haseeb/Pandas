```python
df['Rating'].isnull().values.any()
```

### We can see Ratings has a total of 1474 NaN values, Type has 1 and Android_Ver has 2 NaN values.

- If we have very less number of values we can drop those rows, it won't affect the data drastically

- But if we have a large number of missing values, dropping them wouldn't be an ideal approach. Instead we can fill these null values.

- There are different methods, to fill these values and it completely depends on the type of data we are dealing with.
    - Fill missing values with 0.
    - Fill them with either maximum of the exisitng values or maybe average.

- In this data, we can fill the Ratings with average Ratings grouped by **Category** of the app.

```python
df['Rating'] = df.groupby('Category')['Rating'].transform(lambda x : x.fillna(x.mean()))
```

### Useful methods on single column

```python
def last_four(num):
    return str(num)[-4:]
```

```python
df_tips['last_four_cc'] = df_tips['CC Number'].apply(last_four)
```

```python
df_tips['total_bill'] = df_tips['total_bill'].apply(lambda x : x*2)
```

```python
df_tips.head()
```

### Useful methods on multiple columns

```python
def quality_bill(total_bill,tip):
    if(tip/total_bill > 0.25):
        return 'Generous'
    return 'Other'
```

```python
quality_bill(8.77,2)
```

```python
df_tips['Quality_bill'] = df_tips[['total_bill','tip']].apply(lambda df_tips : quality_bill(df_tips['total_bill'],df_tips['tip']),axis=1)
```

```python
# np.vectorize() awares the function about the DataFrame.
df_tips['Quality']  = np.vectorize(quality_bill)(df_tips['total_bill'],df_tips['tip'])
```

## **MERGE**

```python
# MERGE is analogous to JOIN in SQL
# Three ways to merge tables : Inner, Outer and (left/right)
```

```python
registrations = pd.DataFrame({"reg_id":[1,2,3,4],"Name":['Andrew','Blob','Casie',"Daniel"]})
logins = pd.DataFrame({"login_id":[1,2,3,4],"Name":['Cassandra','Daniel','Xavier',"Blob"]})
```

```python
print(registrations)
print(logins)
```

```python
pd.merge(registrations,logins,how='inner',on='Name')
```

```python
pd.merge(registrations,logins,how='left',on='Name')
```

```python
pd.merge(registrations,logins,how='right',on='Name')
```

```python
pd.merge(registrations,logins,how='outer',on='Name')
```

```python
registrations =  registrations.set_index('Name')
```

```python
# If we want to merge with Index of one table with column of another table
pd.merge(left=registrations,right=logins,how='inner',left_index=True,right_on='Name')
```

```python
registrations = registrations.reset_index()
```

```python
 # If column names are different on which we have to apply the merge
registrations.columns = ['reg_name',"reg_id"]
```

```python
registrations
```

```python
pd.merge(registrations,logins,how='inner',left_on='reg_name',right_on='Name')
```
